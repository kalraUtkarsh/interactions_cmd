```python
import matplotlib.pyplot as plt

fl = open('pipi.txt','r').readlines()
x = []
y = []
for ind, linr in enumerate(fl):
    x.append(ind)
    y.append(float(linr.split(',')[2]))
```

```python
plt.plot(x,y)
plt.title('Theta Plot')
plt.xlabel('File Number')
plt.ylabel('Theta Angle')
plt.show()
```
